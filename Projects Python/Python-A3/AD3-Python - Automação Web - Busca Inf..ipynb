```python
# Passo 1- Pegar cotação Dolar
from selenium import webdriver
from selenium.webdriver.common.keys import Keys

  # Abrir navegador
navegador = webdriver.Chrome()
  # Entrar no google
navegador.get('https://www.google.com.br/')
  # Pesquisar cotação
navegador.find_element('xpath',
                       '/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input').send_keys('Cotacao dolar')   
  # pegar valor dolar
navegador.find_element('xpath',
                       '/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input').send_keys(Keys.ENTER)

  # Capturar valor Cotação
cotacao_dolar=navegador.find_element('xpath',
                                     '//*[@id="knowledge-currency__updatable-data-column"]/div[1]/div[2]/span[1]').get_attribute('data-value')

print(cotacao_dolar)

 # passo 2- Pegar cotação Euro
    
    # Abrir navegador
navegador = webdriver.Chrome()
  # Entrar no google
navegador.get('https://www.google.com.br/')
  # Pesquisar cotação
navegador.find_element('xpath',
                       '/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input').send_keys('Cotacao euro')   
  # pegar valor Euro
navegador.find_element('xpath',
                       '/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input').send_keys(Keys.ENTER)

  # Capturar valor Cotação
cotacao_euro=navegador.find_element('xpath',
                                     '//*[@id="knowledge-currency__updatable-data-column"]/div[1]/div[2]/span[1]').get_attribute('data-value')

print(cotacao_euro)

# passo 3- Pegar cotação Ouro

  # Entrar no google
navegador.get('https://www.melhorcambio.com/ouro-hoje')

  # Capturar valor Cotação
cotacao_ouro=navegador.find_element('xpath','//*[@id="comercial"]').get_attribute('value')
cotacao_ouro = cotacao_ouro.replace(',','.')

print(cotacao_ouro)

navegador.quit()
 
```

```python
# Passo 4- Atualizar base de dados 
import pandas as pd

# importar tabela de dados

tabela = pd.read_excel('Produtos.xlsx')
display(tabela)
```

```python
# Passo 5- Recalcular os precos

# 1 - atualizar cotação
tabela.loc[tabela['Moeda']=='Dólar','Cotação']= float(cotacao_dolar)
tabela.loc[tabela['Moeda']== 'Euro','Cotação']= float(cotacao_euro)
tabela.loc[tabela['Moeda']== 'Ouro','Cotação']= float(cotacao_ouro)

# 2- ajustar preço de compra = Preco original * cotacao 
tabela['Preço de Compra']= tabela['Preço Original'] * tabela['Cotação']

# 3- Ajustar preço de venda = Preço de compra * margem
tabela['Preço de Venda']= tabela['Preço de Compra'] * tabela['Margem']


display(tabela)

```

```python
# Passo 6- Exportar Base de dados
tabela.to_excel('Produtos_Novo.xlsx', index=False)
```
