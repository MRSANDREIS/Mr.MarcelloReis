```python
import pyautogui   # Autmatiza processos
import pyperclip  # ler Caracteres especiais 
import time       # tempo espera para carregamento
import datetime
pyautogui.PAUSE=1
#pyautogui.click--> Clicar com mouse
#pyautogui.write--> Escreve um texto
#pyautogui.press--> pressiona um botao

#1- Passo - Entrar no sistema (entrar no link)
pyautogui.hotkey('ctrl','t')
pyperclip.copy("https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga?usp=sharing")
pyautogui.hotkey('Ctrl','v')
pyautogui.press('ENTER')
# Aguardando site carregar
time.sleep(5)
#2- Passo - Navegar no sistema  e encontrar a base de dados (entra na pasta exportar)
pyautogui.click(x=335, y=255,clicks=2)
time.sleep(2)

#3- Passo - Download da base de dados

pyautogui.click(x=366, y=260)  # clicando no arquivo

pyautogui.click(x=1158, y=162) # Selecionando local download

pyautogui.click(x=937, y=561)  # Realizando download
time.sleep(5)
```

```python
#4- Buscar base de dados e Calcular os indicadores (Faturamento,qtidade de produtos)
import pandas as pd
tabela= pd.read_excel(r'C:\Users\mrmar\Downloads\Vendas - Dez.xlsx')

# Calulando indicadores (Quantidade e faturamento)

# Qty_produtos= soma da coluna quantidade
qty_produtos = tabela['Quantidade'].sum()
print(qty_produtos)

# faturamento= soma da coluna valor final
total_faturamento= tabela['Valor Final'].sum()
print(total_faturamento)

display(tabela)
```

```python
#5- Passo - Entra email
pyautogui.hotkey('ctrl','t')  # Abriu nova aba
pyperclip.copy("mail.google.com/mail/u/0/#inbox") # copiando link
pyautogui.hotkey('Ctrl','v') # Colando link
pyautogui.press('ENTER')     # entrando no email
time.sleep(7)

# Passo ao Clickar no escrever
pyautogui.click(x=87, y=173) # Clicando botao escrever ou (+)
time.sleep(2)
# Escrever destinatario
pyautogui.write("mr.marcelloreis@gmail.com")
pyautogui.press('tab')    # Confirmando email
pyautogui.press('tab')    # Passando para adicionar assunto

#Escrever Assunto
pyperclip.copy("Relatório de Vendas")
pyautogui.hotkey('Ctrl',"V")
time.sleep(1)
pyautogui.press('tab')    # Passando para corpo do email

# Escrevendo msg corpo do email
texto=f'''

Prezados, Bom dia!

O faturamento de ontem , foi de : R${total_faturamento:.2f}
A quantidade de produtos foi de : {qty_produtos:,}

Abs!
Marcello Reis
'''
    
pyperclip.copy(texto)
pyautogui.hotkey('Ctrl','v')

#Enviando email
pyautogui.hotkey("Ctrl",'Enter')

```
