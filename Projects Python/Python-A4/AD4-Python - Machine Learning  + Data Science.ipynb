```python
# Importar Base de dados
import pandas as pd
tabela= pd.read_csv(r"C:\Users\mrmar\Documents\PythonDSA\Python\Marcello\PytonHashtag\Python-A4\advertising.csv")

display(tabela) # Mostra Base_dados
```

```python
# Analise Exploratoria
# Vamos tentar visualizar como as inf. de cada item estao distribuidas
# vamos ver a correlação entre cada um dos itens
import seaborn as sms
import matplotlib.pyplot as plt

display(tabela.corr())

# criar graficos
sms.heatmap(tabela.corr(),cmap="Wistia", annot=True)

# Exibir grafico
plt.show()
```

```python
# Com isso, podemos partir para a preparacao dos dados para treinarmos o modelo de Machine Learning
# y = quem voce quer prever
# x = é o resto todo

x= tabela[['TV','Radio','Jornal']]  

y= tabela['Vendas']

# Tabela será dividida em 4 partes pelo sklearn
from sklearn.model_selection import train_test_split
x_treino, x_teste, y_treino, y_teste = train_test_split(x,y)
```

```python
# Temos um problema de regressão - vamos escolher os modelos que vamos usar
# --> Regressao linear
# --> RandomForest(Arvore de decisão)
# Vamos importar inteligencia artificial 

from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor

# Cria a inteligencia

modelo_regressaolinear = LinearRegression()
modelo_arvoredecisao = RandomForestRegressor()

# Vamos Treinar a inteligencia Artificial

modelo_regressaolinear.fit(x_treino,y_treino)
modelo_arvoredecisao.fit(x_treino,y_treino)
 
```

```python
# Teste da AI avaliação do modelo
# Vamos usar o R2--> diz o % que o nosso modelo consegue explicar o que acontece
# Previsao dos teste

previsao__regressaolinear = modelo_regressaolinear.predict(x_teste)
previsao_arvoredecisao = modelo_arvoredecisao.predict(x_teste)

from sklearn.metrics import r2_score

print(r2_score(y_teste,previsao__regressaolinear))

print(r2_score(y_teste,previsao_arvoredecisao))
```

```python
# Criando uma tabela auxiliar
tabela_auxiliar = pd.DataFrame()
tabela_auxiliar ["y_teste"]= y_teste
tabela_auxiliar ["arvoredecisao"]=previsao_arvoredecisao
tabela_auxiliar ['regressaolinear']= previsao__regressaolinear

plt.figure(figsize=(15,8))
sms.lineplot(data=tabela_auxiliar)
plt.show()
```
